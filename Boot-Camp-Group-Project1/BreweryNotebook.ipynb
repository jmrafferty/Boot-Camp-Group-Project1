```python
import json
import pandas as pd
from pprint import pprint
import datetime

file = "Brewery_DistilleryData.csv"
```

```python
brewerydf = pd.read_csv(file)
brewerydf.fillna(0, inplace = True)
brewerydf.head()

#reformat date column to eliminate timestamps
brewerydf['Issue Date'] = pd.to_datetime(brewerydf['Issue Date'])

brewerydf.head(100)
```

```python
#cleaning the data

startdate = "01/01/2009"
enddate = "12/31/2009"
daterange = [startdate,enddate]
activelist = brewerydf.loc[(brewerydf["Business Status"] == "Active") 
                        & (brewerydf["Permit Status"] == "Active")
                        & (brewerydf["Issue Date"] < daterange[1])
                        & (brewerydf["Issue Date"] > daterange[0])
                        & (brewerydf["Permit Number"].str.contains("AE", regex=False) == True)
                          ]
activelist.tail(100)
```
