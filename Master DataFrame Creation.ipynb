```python
# Dependencies
import pandas as pd
import re
```

```python
filelist=["../Project 1/Zip_Zhvi_AllHomes.csv",
          "Zip_MedianListingPricePerSqft_AllHomes.csv",
          "Zip_MedianListingPrice_AllHomes.csv",
         "Zip_ZriPerSqft_AllHomes.csv"]

dfnames=["zhviAll",
         "Mediansqft",
         "Median",
         "Mediansqftrent"] #Make sure this list is same length as 'filelist'
```

```python
dfs=[]
for i in range(len(filelist)): #read in each file and drop unnessesary columns
    df=pd.read_csv(filelist[i], encoding="ISO-8859-1")
    try:
        df.drop(columns=['RegionID','SizeRank'])
    except:
        print(f"No RegionID Column in {dfnames[i]}") #error message if the columns to be dropped don't exist
    dfs.append(df.loc[df["State"]=="NC"]) #Filter files by State
    print(f'{"{0:.0%}".format((i+1)/len(filelist))} Complete',end='\r') #'end=\r' overwrites the existing line
```

```python
for i in range(len(dfs)):  #iterate through each list and add suffixes to prepare for merge 
    dfs[i].columns = [str(col) + f'_{dfnames[i]}' if col != "RegionName" else col for col in dfs[i].columns]
```

```python
MergedDF=dfs[0].merge(dfs[1], on='RegionName', how="outer").merge(dfs[2], on='RegionName', how="outer").merge(dfs[3], on='RegionName', how="outer")
MergedDF
```

```python
MergedDF.to_csv('../Project 1/MergedDF.csv') #write to a csv and save locally
```
