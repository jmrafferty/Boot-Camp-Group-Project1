```python
import json
import pandas as pd
import numpy as np
from pprint import pprint
import datetime
import re

file = "Brewery_DistilleryData.csv"
```

```python
brewerydf = pd.read_csv(file)
brewerydf.fillna(0, inplace = True)
brewerydf.head()

#reformat date column to eliminate timestamps
brewerydf['Issue Date'] = pd.to_datetime(brewerydf['Issue Date'])
brewerydf["OpenYear"] = pd.DatetimeIndex(brewerydf["Issue Date"]).year
brewerydf["OpenMonth"] = pd.DatetimeIndex(brewerydf["Issue Date"]).month
brewerydf.tail(30)
```

```python
#cleaning the data
activelist = brewerydf.loc[(brewerydf["Business Status"] == "Active") 
                & (brewerydf["Permit Status"] == "Active")
                #& (brewerydf["OpenYear"] == year)
                & (brewerydf["Permit Number"].str.endswith("BM"))
                  ]
outputdf = activelist[["Trade Name","Zip","OpenYear","OpenMonth"]].copy()
outputdf.reset_index(inplace=True)
outputdf.to_csv("")
```
